## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM master
"""
pd.read_sql_query(QUERY,conn).hist()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """
SELECT Age
FROM master
"""
pd.read_sql_query(QUERY,conn).plot(kind="box")
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, AVG(WorkWeekHrs) as hours
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn).plot(x="Age", y= "hours")
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, AVG(WorkWeekHrs) as Work, AVG(CodeRevHrs) as Code
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn).plot.scatter(x="Work", y= "Code", s="Age")
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as cant
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by cant DESC
"""

pd.read_sql_query(QUERY,conn).set_index("DatabaseDesireNextYear").head(5).plot.pie(y="cant",autopct='%1.0f%%')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""

data = pd.read_sql_query(QUERY,conn).head(5)
data = data.groupby("Age").median()
data.plot(kind="bar", stacked=True)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 25 AND 30
"""

data = pd.read_sql_query(QUERY,conn).head(5)
data = data.groupby("Age").median()
data.plot()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, Count(*)
FROM master
Group by MainBranch
"""

pd.read_sql_query(QUERY,conn).set_index("MainBranch").plot(kind="barh")
```
